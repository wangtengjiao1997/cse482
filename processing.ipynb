```python
import pandas as pd
import datetime
data_ori = pd.read_csv("CARGSP.csv")
data_ori.head()
```

```python
data_gdp = pd.read_csv("CARGSP.csv")
z= 0
for x in data_gdp["DATE"]:
    data_gdp["DATE"][z] = x[0:4]
    z = z+1
data_gdp.head()
```

```python
data_exchange = pd.read_csv("DEXUSEU.csv")
data_exchange
```

```python
data_stock = pd.read_csv("AAPL.csv")
data_stock.head()
```

```python
data_stock["DATE"] = data_stock["Date"]
data_stock = data_stock.drop(columns="Date")
data_stock.head()
```

```python
data_new = {"DATE":[],"GDP":[]}
for date in data_exchange["DATE"]:
    for index , value in data_gdp.iterrows():
        if value[0] == date[:4]:           
            data_new["DATE"].append(date)
            data_new["GDP"].append(value[1])
dataf_new = pd.DataFrame(data_new)            
dataf_new
```

```python
merge1 = pd.merge(data_exchange, data_stock, left_on = 'DATE', right_on = 'DATE', how = 'inner')
merge1
```

```python
merge2 = pd.merge(dataf_new, merge1, left_on = 'DATE', right_on = 'DATE', how = 'inner')
merge2
```

```python
data_dow = pd.read_csv("DJIA.csv")
data_dow.head()
```

```python
merge3 = pd.merge(data_dow, merge2, left_on = 'DATE', right_on = 'DATE', how = 'inner')
merge3
```

```python
data_nas = pd.read_csv("NASDAQ100.csv")

merge3 = pd.merge(data_nas, merge3, left_on = 'DATE', right_on = 'DATE', how = 'inner')
merge3.info()
```

```python
z = 0

merge4 = {"DATE":[],"Predict":[]}
while z < len(merge3)-30:
    Goes = merge3['Close'][z+30]-merge3['Close'][z]    
    if Goes >= 0:
        inc = 'Increase'
        merge4['DATE'].append(merge3["DATE"][z])
        merge4['Predict'].append(inc)
    else:
        dec = 'Decrease'
        merge4['DATE'].append(merge3["DATE"][z])
        merge4['Predict'].append(dec)
    z+=1
merge4_n = pd.DataFrame(merge4) 
final_merge = pd.merge(merge3, merge4_n,left_on = 'DATE', right_on = 'DATE', how = 'inner')
final_data = final_merge.drop(['Open'],axis=1)
final_data = final_data.drop(['High'],axis=1)
final_data = final_data.drop(['Low'],axis=1)
final_data = final_data.drop(['Adj Close'],axis=1)
final_data["DJIA"] = pd.to_numeric(final_data["DJIA"])
v = 0
while v < len(final_data["DEXUSEU"]):
    if(final_data["DEXUSEU"][v]) == ".":
        final_data["DEXUSEU"][v] = final_data["DEXUSEU"][v+1]
    v+=1
merge3["DEXUSEU"] = pd.to_numeric(final_data["DEXUSEU"])
merge3["NASDAQ100"] = pd.to_numeric(final_data["NASDAQ100"])
final_data
```

```python
from sklearn.model_selection import train_test_split
Y = final_data['Predict']
X = final_data.drop('Predict', axis=1)
X = X.drop("DATE", axis =1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.8, random_state=1)

from sklearn import tree
from sklearn.metrics import accuracy_score

clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
tree_acc = accuracy_score(Y_test, Y_pred)
print('Accuracy',tree_acc)

from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline

numNeighbors = [1, 5, 10, 15, 20, 25, 30]
knn_acc = []

for nn in numNeighbors:
    clf = KNeighborsClassifier(n_neighbors=nn)
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_test)
    knn_acc.append(accuracy_score(Y_test, Y_pred))
    
plt.plot(numNeighbors, knn_acc, 'ro-')
plt.xlabel('Number of neighbors')
plt.ylabel('Test accuracy')
```
